```python
import pandas as pd
```

```python
completion_positive = client.chat.completions.create(
  model="gpt-4-1106-preview",
  messages=[
    {"role": "system", "content": "You are an Engineer. You are trying to create AI for Thai language, but the dataset resources for Thai language is not enough."},
    {"role": "user", "content":   "Generate mixed 70 long Thai sentences in positive tone. Imitate the comments in the internet. Each sentence should sound natural and diverse. Don't use too much exclamation marks. Use a variety of vocabulary and grammar structures. Show only the generated sentences. Translation, or introduction are unnecessary."},
  ]
)
```

```python
completion_neutral = client.chat.completions.create(
  model="gpt-4-1106-preview",
  messages=[
    {"role": "system", "content": "You are an Engineer. You are trying to create AI for Thai language, but the dataset resources for Thai language is not enough."},
    {"role": "user", "content":   "Generate mixed 70 long Thai sentences in neutral tone. Imitate the comments in the internet. Each sentence should sound natural and diverse. Use a variety of vocabulary and grammar structures. Show only the generated sentences. Translation, or introduction are unnecessary."},
  ]
)
```

```python
completion_negative = client.chat.completions.create(
  model="gpt-4-1106-preview",
  messages=[
    {"role": "system", "content": "You are an Engineer. You are trying to create AI for Thai language, but the dataset resources for Thai language is not enough."},
    {"role": "user", "content":   "Generate mixed 70 long Thai sentences in negative tone. Imitate the comments in the internet. Each sentence should sound natural and diverse. Use a variety of vocabulary and grammar structures. Show only the generated sentences. Translation, or introduction are unnecessary."},
  ]
)
```

```python
print(completion_positive.choices[0].message)
print(completion_neutral.choices[0].message)
print(completion_negative.choices[0].message)
positive_output_msg = completion_positive.choices[0].message.content
neutral_output_msg = completion_neutral.choices[0].message.content
negative_output_msg = completion_negative.choices[0].message.content
```

```python
splitted_msg_pos = positive_output_msg.split("\n")
splitted_msg_neu = neutral_output_msg.split("\n")
splitted_msg_neg = negative_output_msg.split("\n")
# split the output_msg into bullet points
pos_bullet_points = [point.strip() for point in splitted_msg_pos if point.strip()]
neu_bullet_points = [point.strip() for point in splitted_msg_neu if point.strip()]
neg_bullet_points = [point.strip() for point in splitted_msg_neg if point.strip()]
```

```python
plain_dict_pos = {point.split('. ')[1]: "POS" for point in pos_bullet_points if '. ' in point and len(point.split('. ')) > 1}
plain_dict_neu = {point.split('. ')[1]: "NEU" for point in neu_bullet_points if '. ' in point and len(point.split('. ')) > 1}
plain_dict_neg = {point.split('. ')[1]: "NEG" for point in neg_bullet_points if '. ' in point and len(point.split('. ')) > 1}
```

```python
# create a dataframe from the dictionary
df_pos = pd.DataFrame.from_dict(plain_dict_pos, orient='index', columns=['label'])
df_neu = pd.DataFrame.from_dict(plain_dict_neu, orient='index', columns=['label'])
df_neg = pd.DataFrame.from_dict(plain_dict_neg, orient='index', columns=['label'])

df_pos.reset_index(inplace=True)
df_pos.rename(columns={'index': 'Text'}, inplace=True)

df_neu.reset_index(inplace=True)
df_neu.rename(columns={'index': 'Text'}, inplace=True)

df_neg.reset_index(inplace=True)
df_neg.rename(columns={'index': 'Text'}, inplace=True)

# display the dataframe
# df_pos.style.set_properties(**{'text-align': 'left'})
# df_neu.style.set_properties(**{'text-align': 'left'})
# df_neg.style.set_properties(**{'text-align': 'left'})
# splitted_msg
```
